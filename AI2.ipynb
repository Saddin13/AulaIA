```python
import numpy as np

#1- Define o problema (Fenotipo)
#PROBLEMA : OPTIMIZAR ITENS NA MOCHILA

itens = [
    {"nome": "Notebook" , "valor" : 2000, "peso": 3},
    {"nome": "Mouse"    , "valor" : 100 , "peso": 1},
    {"nome": "Teclado"  , "valor" : 300 , "peso": 1},
    {"nome": "Monitor"  , "valor" : 1500, "peso": 4},
    {"nome": "Cadeira"  , "valor" : 1000, "peso": 10},
]

PesoLimite = 10

#. parametros

Tamanho_Populacao = 10
Numero_Genes = len(itens)

def criar_indiviuo():
  """Gera um indiviuo aleatorio (Genotipo)"""
  return np.random.randint(2, size=Numero_Genes).tolist()

def fitness(cromossomo):
  """
  Avalia a solução (Baseado em Linden: Penalição)
  Se o peso exceder o limite, o fitness é reduzido
  """
  peso_total = 0
  valor_total = 0
  for i in range(Numero_Genes):
    if cromossomo[i] == 1:
      peso_total += itens[i]["peso"]
      valor_total += itens[i]["valor"]
  if peso_total > PesoLimite:
    return 1 # penalização: fitness minimo para solucoes inviaveis
  return valor_total

#3. EXECUÇÃO: CRIANDO A POPULACAO
populacao = [criar_indiviuo() for _ in range(Tamanho_Populacao)]
print(populacao)
```
