```python
import random

def valoresaleatorios():
  T = 50
  numeros = []
  for i in range(T):
    numeros.append(random.randint(-500,500))
  return numeros # Moved return outside the loop

def maiorvalor(lista):
  maior = lista[0]
  for i in range(1, len(lista)):
    if lista[i] > maior:
      maior = lista[i]
  return maior

Vmax = None
numeros = valoresaleatorios()
Vmax = maiorvalor(numeros)

for i in range(len(numeros)):
  print(str(numeros[i]) + ", ", end="") # Fixed SyntaxError
print ("\nValor Maximo: " + str(Vmax))
```
