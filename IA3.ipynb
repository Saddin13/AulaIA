```python
def PesoMateria(materia):
  if(materia["Area"]) == "Exatas":
    return 0.6
  elif(materia["Area"]) == "Humanas":
    return 0.3
  elif(materia["Area"]) == "Biologicas":
    return 0.1
  else:
    return 0
```

```python
# Re-definindo QMaterias, CriarAlunos e VericarGrade para garantir que as versões mais recentes sejam usadas.
# Assume-se que a lista 'disciplinas' e o módulo 'random' já foram carregados em células anteriores.

# Definição de QMaterias
QAlunos = 100
QMaterias = len(disciplinas)

# Definição CORRIGIDA da função CriarAlunos
def CriarAlunos():
  """Gera a grade de 1 aluno ALEATORIAMENTE"""
  opcoes = []
  for i in range(QMaterias):
    opcoes.append(random.randint(0,1))
  return opcoes # 'return' está FORA do loop

# Definição da função VericarGrade
def VericarGrade(grade):
  totalCredito = 0
  totalPeso = 0
  materiasSelecionadasCount = 0
  for i in range(QMaterias):
    if grade[i] == 1:
      totalCredito = totalCredito + disciplinas[i]["Creditos"]
      totalPeso = totalPeso + PesoMateria(disciplinas[i])
      materiasSelecionadasCount = materiasSelecionadasCount + 1

  print(f"Total de Creditos: {totalCredito}")
  if materiasSelecionadasCount > 0:
    mediaPeso = totalPeso / materiasSelecionadasCount
    print(f"Média Ponderada da Grade: {mediaPeso:.2f}")
  else:
    print("Nenhuma matéria selecionada na grade.")

  # Adiciona a verificação de créditos
  if 140 <= totalCredito <= 150:
    print("Status da Grade: Válida (Créditos entre 140 e 150)")
  else:
    print("Status da Grade: Inválida (Créditos fora da faixa de 140-150)")

# Chamada da função para verificar a grade do aluno recém-criado
VericarGrade(CriarAlunos())
```

```python
# Gerando e verificando grades para cada um dos QAlunos
for aluno_id in range(QAlunos):
  print(f"\nVerificando grade para Aluno {aluno_id + 1}:")
  grade_aluno = CriarAlunos()
  VericarGrade(grade_aluno)
```

```python
def EncontrarMelhorAluno(QAlunos):
  melhor_aluno_id = -1
  maior_media_peso = -1.0
  melhor_grade = []
  melhor_total_credito = 0

  for aluno_id in range(QAlunos):
    grade_aluno = CriarAlunos()
    totalCredito = 0
    totalPeso = 0
    materiasSelecionadasCount = 0

    for i in range(QMaterias):
      if grade_aluno[i] == 1:
        totalCredito += disciplinas[i]["Creditos"]
        totalPeso += PesoMateria(disciplinas[i])
        materiasSelecionadasCount += 1

    # Aplica a verificação de créditos: se fora da faixa 140-150, pula para o próximo aluno
    if not (0 <= totalCredito <= 150):
      continue # Pula esta grade se os créditos não estiverem na faixa válida

    if materiasSelecionadasCount > 0:
      mediaPeso = totalPeso / materiasSelecionadasCount
    else:
      # Se chegou aqui com 0 matérias selecionadas mas créditos na faixa, algo está errado.
      # Ou o totalCredito seria 0, o que faria a condição acima ser verdadeira.
      # Então, para este cenário, um mediaPeso de 0 é apropriado.
      mediaPeso = 0

    if mediaPeso > maior_media_peso:
      maior_media_peso = mediaPeso
      melhor_aluno_id = aluno_id
      melhor_grade = grade_aluno
      melhor_total_credito = totalCredito

  print(f"\n--- Melhor Aluno Encontrado ---")
  print(f"ID do Aluno: {melhor_aluno_id + 1}")
  print(f"Média Ponderada da Grade: {maior_media_peso:.2f}")
  print(f"Total de Creditos: {melhor_total_credito}")
  print(f"Grade Selecionada (1=Sim, 0=Não): {melhor_grade}")

# Chamada da função para encontrar o melhor aluno
EncontrarMelhorAluno(QAlunos)
```
